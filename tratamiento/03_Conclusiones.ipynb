# Conclusión General del Análisis Presupuestario y Recaudación
Tras analizar en detalle los datos de los valores previstos, recaudados y sus diferencias, junto con insights específicos sobre las categorías económicas y los ministerios, se destacan los siguientes puntos clave:

## 1. Tendencias Generales en la Planificación y Ejecución Presupuestaria
### Estabilidad y Crecimiento a Largo Plazo
- Tanto los valores previstos como los recaudados muestran un crecimiento general en el periodo analizado (2013-2021), lo que refleja una expansión de los presupuestos y la recaudación gubernamental en términos absolutos.
- Sin embargo, persisten diferencias notables entre lo previsto y lo recaudado, con déficits predominantes en la mayoría de los años. Estos déficits reflejan problemas estructurales en la planificación presupuestaria y la capacidad de recaudación.

### Años Clave: Crisis y Recuperación
- 2017 y 2020 destacan como años con las mayores discrepancias negativas, atribuibles a eventos extraordinarios como cambios económicos o la pandemia de COVID-19.
- 2020 presenta un caso atípico, con un déficit en las Receitas Correntes (-251,36 billones) y un superávit en las Receitas de Capital (305,66 billones). Esto podría relacionarse con políticas de financiamiento y estímulo implementadas en respuesta a la crisis sanitaria.
- 2021 muestra una recuperación con un superávit en las Receitas Correntes (70,47 billones), sugiriendo un repunte económico y mayor eficiencia en la recaudación.



## 2. Insights por Categorías de Ingresos
### Receitas Correntes
- Presentan un comportamiento más estable, con déficits que disminuyen en años recientes y superávits en 2018, 2019, y 2021.
- La recuperación en 2021 destaca como un punto positivo, posiblemente resultado de una mejora en las políticas de recaudación y mayor control fiscal.
### Receitas de Capital
- Muestran mayor volatilidad, con déficits significativos en años como 2017 y 2021.
- El superávit excepcional en 2020 indica medidas extraordinarias para manejar la pandemia, como financiamiento externo o ingresos de capital no recurrentes.



## 3. Desempeño por Ministerios y Fuentes de Ingresos
### Disparidades Entre Ministerios
- Ministerios como la Controladoria-Geral da União y la Advocacia-Geral da União presentan bajos niveles de recaudación respecto a sus presupuestos, lo que indica problemas en la planificación o ejecución de ingresos.
- Por otro lado, algunos ministerios como el Ministério de infraestructura, mujeres e iguladad y pesca e aquicultura lograron porcentajes de cumplimiento superiores al 100%, indicando proyecciones más precisas o ingresos no previstos.



### Fuentes Destacadas
- Ingresos como "Amortizações de Empréstimos" y "Receitas Patrimoniais" han superado significativamente las previsiones en algunos años, lo que sugiere una buena gestión o ingresos adicionales inesperados.
- Categorías como "Alienação de Bens" y "Transferências de Capital" muestran un desempeño consistentemente bajo, requiriendo análisis para identificar barreras en su ejecución.
### Outliers y Anomalías
Valores negativos o cero en categorías como "Sem Informação" y "Receitas Correntes - a Classificar" indican errores contables o falta de claridad en la clasificación de ingresos, lo que afecta la confiabilidad del análisis.




## 4. Retos y Áreas de Mejora
### Proyecciones Presupuestarias
- Los déficits recurrentes resaltan la necesidad de mejorar la precisión en las estimaciones de ingresos. Esto incluye la incorporación de análisis más realistas basados en datos históricos y factores externos como fluctuaciones económicas o cambios regulatorios.
### Gestión de Ingresos
- La disparidad en los porcentajes realizados entre ministerios sugiere ineficiencias en la recaudación y ejecución de políticas fiscales. Es necesario un monitoreo más riguroso y la implementación de mejores prácticas administrativas.
### Registro y Transparencia
- La falta de datos en categorías como el "Valor Lanzado" y las inconsistencias en algunas fuentes de ingresos subrayan la necesidad de mejorar los sistemas de registro y la rendición de cuentas para garantizar datos más confiables y útiles para la toma de decisiones.

## 5. Conclusiones Finales
### Avances Notables: 
Los superávits recientes y el crecimiento en los valores recaudados muestran progresos en ciertas áreas, con ejemplos de buena planificación en fuentes como "Receitas Patrimoniais" y ministerios como el Ministério do Desenvolvimento Regional.

### Desafíos Persistentes: 
Las diferencias significativas entre lo previsto y lo recaudado, especialmente en categorías como "Receitas de Capital", reflejan desafíos estructurales y la necesidad de fortalecer las capacidades de planificación y ejecución financiera.

## Recomendaciones Clave:

- Implementar métodos de proyección más robustos y basados en datos para mejorar la precisión de los presupuestos.
- Fortalecer los mecanismos de recaudación en ministerios con desempeño bajo.
- Investigar y corregir inconsistencias en los registros para mejorar la calidad y transparencia de los datos financieros.

#### En resumen, aunque se observan señales de recuperación y mejoras en ciertas áreas, es esencial abordar las brechas existentes y optimizar los procesos de planificación y recaudación para asegurar una gestión financiera más eficiente y equitativa.
